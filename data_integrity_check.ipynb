```python
import pandas as pd

hard_df = pd.read_csv("hard_complete.csv")
easy_df = pd.read_csv("easy_complete.csv")

easy_df.head()
```

```python
from RNAFoldAssess.utils import *
```

```python
def _count_Ak(series, k: int):
    return (
        series.fillna("")                # handle missing
              .str.upper()               # normalize case
              .map(lambda s: SequenceTools.count_homopolymers(s, k, "A"))
              .astype(int)
    )
hard_df = hard_df.assign(
    **{
        "4merAs": _count_Ak(hard_df["sequence"], 4),
        "5merAs": _count_Ak(hard_df["sequence"], 5),
        "6merAs": _count_Ak(hard_df["sequence"], 6),
    }
)

easy_df = easy_df.assign(
    **{
        "4merAs": _count_Ak(easy_df["sequence"], 4),
        "5merAs": _count_Ak(easy_df["sequence"], 5),
        "6merAs": _count_Ak(easy_df["sequence"], 6),
    }
)
```
